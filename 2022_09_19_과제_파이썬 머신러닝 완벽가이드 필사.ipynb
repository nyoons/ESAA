>**DataFrame 데이터 삭제**
<br/>drop() 메서드 이용.
<br/>원형은 DataFrame.drop(**labels=None**, **axis=0**, index=None, columns=None, level=None, **inplace=False**, errors='raise')
*   axis - axis에 따라 특정 칼럼 또는 행을 드롭.
*   labels - labels에 칼럼 명을 입력하고 axis=1 입력하면 지정된 칼럼 드롭. 여러개의 칼럼을 삭제하고 싶으면 리스트 형태로 입력하여 labels 파라미터로 입력한다.

>**Index 객체**
*   판다스의 Index 객체는 Primary Key와 유사하게 DF, Series의 레코드를 고유하게 식별하는 객체. DF, Series에서 Index 객체만 추출하려면 DataFrame.index 또는 Series.index 속성을 통해 가능하다. 1차원 array로 식별성 데이터를 가지고 있고, ndarray와 유사하게 단일 값 반환 및 슬라이싱도 가능하다.
*   하지만 한 번 만들어진 DataFrame 및 Series의 Index 객체는 함부로 변경할 수 없다. ex) indexes[0]=5 는 불가
*   Index는 오직 식별용이라 연산함수를 적용할 때는 제외한다.
*   DataFrame.**reset_index**(inplace=False) 메서드를 수행하면 새롭게 인덱스를 연속 숫자 형으로 할당하며 기존 인덱스는 'index'라는 새로운 칼럼명으로 추가한다. 인덱스가 연속된 int숫자형 데이터가 아닐 경우, 다시 이를 연속 int숫자형 데이터로 만들때 사용한다. 

>**데이터 셀렉션 및 필터링**
<br/>혼동하기 쉽다. 넘파이에선 [ ] 연산자 내 단일 값 추출, 슬라이싱, 팬시 인덱싱, 불린 인덱싱 수행. **판다스에서는 ix[ ], iloc[ ], loc[ ] 연산자**로 수행.
1. DataFrame의 [ ] 연산자
<br/>**DataFrame에서 [ ] 안에는 칼럼명(문자)** 또는 인덱스로 변환 가능한 표현식(그냥 숫자 하나는 안됨. 슬라이싱 또는 titanic_df[titanic_df['Pclass']==3].head(3)같은 불린 인덱싱)이 들어간다.
2. DataFrame ix[ ] 연산자
 * ix[0, 'Pclass']처럼 행 위치 지정으로 인덱스값 0, 열 위치 지정으로 'Pclass' 칼럼의 원하는 위치의 데이터를 추출할 수 있다. 열 위치 지정은 칼럼명 뿐만 아니라 컬럼의 위치 값 지정도 가능하다. 단일 지정, 슬라이싱, 불린, 팬시 모두 가능하나 현재는 사라질 예정. loc[ ], iloc[ ] 사용 권장
3. 명칭 기반 인덱싱과 위치 기반 인덱싱의 구분
<br/>만약 data_df의 인덱스값이 문자열이아닌 숫자 값(0,1,2...)라면 data_df.ix[0, :]의 행 위치 값 0은 인덱스 값이 0인건지, 첫번째 행 위치인지 헷갈린다. 인덱스가 integer형인 경우 ix[0,1]은 행 위치 값으로 위치 기반 인덱스값이 아니라 **DataFrame의 인덱스값을 사용**하는데, 인덱스값이 0인 key는 찾을 수 없어 오류를 반환한다. 결과적으로 **DataFrame의 인덱스값은 명칭 기반 인덱싱**이라고 간주.
4. DataFrame iloc[ ] 연산자
<br/>iloc[ ] 는 위치 기반 인덱싱만 허용한다. 슬라이싱과 팬시 인덱싱은 제공하나 위치 기반 인덱싱 제약으로 불린 인덱싱은 제공하지 않는다.
5. DataFrame loc[ ] 연산자
<br/>명칭 기반으로 데이터 추출. **슬라이싱 할 때 loc는 종료값까지 포함**해서 출력한다.
6. 불린 인덱싱
<br/>매우 편리. [ ], ix[ ], loc[ ]에서 모두 지원. 하지만 iloc[ ] 는 제외.

>**정렬, Aggregation 함수, GroupBy 적용**
1. DataFrame, Series의 정렬 - sort_values()
<br/>주요 입력 파라미터는 by,ascending, inplace. by=['칼럼명']으로 해당 칼럼으로 정렬할 칼럼 선택. ascending=True는 오름차순으로 정렬(기본). inplace=False는 sort_values()를 호출한 DF는 그대로 유지하며 정렬된 DF를 결과로 반환(기본). 
2. Aggregation 함수 적용
<br/>min(), max(), sum(), count() 같은 aggregation 함수의 적용은 SQL과 유사하나 DF의 경우 DF에서 바로 aggregation을 호출할 경우 모든 칼럼에 해당 aggregation을 적용한다. (ex. titanic_df.count()) 특정 칼럼에만 적용하기 위해서는 DF에 대상 칼럼들만 추출해 aggregation을 적용하면 된다. (titanic_df[['Age', 'Fare']].mean())
3. groupby() 적용
 * 입력 파라미터 by에 칼럼을 입력하면(by='Pclass') 대상 칼럼으로 groupby 된다. DF에 groupby()를 호출하면 DataFrameGroupBy라는 또 다른 형태의 DF를 반환한다.
 * DF에 groupby()를 호출해 반환된 결과에 aggregation 함수를 호출하면 groupby() 대상 칼럼을 제외한 모든 칼럼에 해당 aggregation 함수를 적용한다. [ ] 이용시 칼럼 선택도 가능하긴 함.
 * DF groupby()의 경우 적용하려는 여러 개의 aggregation 함수명을 DataFrameGroupBy 객체의 agg() 내에 인자로 입력해서 사용한다. ex) titanic_df.groupby('Pclass')['Age'].agg([max, min]) -> 조금 복잡
 * agg() 내에 입력 값으로 딕셔너리 형태 사용 가능.

```python
titanic_groupby=titanic_df.groupby(by='Pclass')
print(type(titanic_groupby))
```

```python
titanic_groupby=titanic_df.groupby('Pclass').count()
titanic_groupby
```

```python
titanic_groupby=titanic_df.groupby('Pclass')[['PassengerId', 'Survived']].count()
titanic_groupby
```

```python
titanic_df.groupby('Pclass')['Age'].agg([max,min]) #agg 안의 형태 주의
```

```python
#agg 내 딕셔너리 형태
agg_format={'Age':'max', 'SibSp':'sum', 'Fare':'mean'}
titanic_df.groupby('Pclass').agg(agg_format)
```

>**결손 데이터 처리하기**

>**apply lambda 식으로 데이터 가공**
<br/>판다스는 apply 함수에 lambda 식을 결합해 DataFrame이나 Series의 레코드별로 데이터를 가공하는 기능을 제공한다.
 * lambda 식이란, 파이썬에서 함수형 프로그래밍을 지원하기 위해 만든 것.


```python
#lambda 식이란
def get_square(a):
  return a**2
print('3의 제곱은:', get_square(3))

lambda_square = lambda x:x**2 # : 로 입력 인자와 반환될 입력 인자의 계산을 분리한다.
print('3의 제곱은:', lambda_square(3))

#여러 개의 값을 입력 인자로 사용할 경우
a=[1,2,3]
squares=map(lambda x:x**2, a)
list(squares)
```

<br/>titanic_df['Name_len']=titanic_df['Name'].apply(lambda x:len(x)) 으로 쓰인다.

```python
titanic_df['Child_Adult']=titanic_df['Age'].apply(lambda x: 'Child' if x<=15 else 'Adult')
titanic_df[['Age', 'Child_Adult']].head(8)
```

```python
titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x:'Child' if x<=15 else ('Adult' if x<=60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
```

```python
#나이에 따라 세분화된 분류를 수행하는 함수 생성
def get_category(age):
  cat=''
  if age<=5: cat='Baby'
  elif age<=12:cat='Child'
  elif age<=18:cat='Teenager'
  elif age<=25:cat='Student'
  elif age<=35:cat='Young Adult'
  elif age<=60:cat='Adult'
  else:cat='Elderly'

  return cat

#lambda 식에 위에서 생성한 get_category() 함수를 반환값으로 지정
#get_category(X)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x:get_category(x))
titanic_df[['Age', 'Age_cat']].head()
```
